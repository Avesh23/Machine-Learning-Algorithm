```python
import pandas as pd
```

```python
data = pd.DataFrame({
    "Size":["s","m","l","xl","s","s","s","xl","xl","xl","l","l","l","l","s","m","l","xl","s","s","s","xl","xl","xl","l","l","l","l"]
    })
```

```python
data.head()
```

```python
ord_data=[["s","m","l","xl"]]      #by deault a b c me rehta h lekin apne ko s m l xl me chahiye isliye aapun usko alag se variable me store kiye h 
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
```

```python
oe.fit(data[["Size"]])
```

```python

data["Size_en"] =oe.transform(data[["Size"]])
```

## ordinal encoding using map 

```python
ord_data1 = {
    "s":0,"m":1,"l":2,"xl":3
}
```

```python
ord_data1
```

```python
data["size_en_map"]= data["Size"].map(ord_data1)
```
