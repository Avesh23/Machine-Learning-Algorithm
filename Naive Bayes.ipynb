```python
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"D:\Data_Analytics\Machine learning\Datasets\placement-dataset.csv") 
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
sns.scatterplot(x="cgpa",y="iq",data=data,hue="placement")
```

```python
x=data[["cgpa","iq"]]
y=data["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
sns.kdeplot(data=data["iq"])
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

## GaussianNB

```python
gnb = GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100,gnb.score(x_train,y_train)*100
```

## MultinomialNB

```python
mnb = MultinomialNB()
```

```python
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)*100,mnb.score(x_train,y_train)*100
```

## BernoulliNB

```python
bnb = BernoulliNB()
```

```python
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test)*100,bnb.score(x_train,y_train)*100
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
plt.show()
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.show()
```
