```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("D:/Data_Analytics/Machine learning/archive/loan_data_set.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
sns.distplot(data["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms =  MinMaxScaler()
```

```python
ms.fit(data[["CoapplicantIncome"]])
```

```python
data["CoapplicantIncome_min"]= ms.transform(data[["CoapplicantIncome"]])
```

```python
data.head()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.histplot(data["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.histplot(data["CoapplicantIncome_min"])
plt.show()
```
